```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=1400,
    height=700,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_scaled_data_nparr = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h",
                            "price_change_percentage_7d", "price_change_percentage_14d", "price_change_percentage_30d",
                            "price_change_percentage_60d", "price_change_percentage_200d", "price_change_percentage_1y"]])

market_scaled_data_nparr[0:41]
```

```python
# Create a DataFrame with the scaled data

df_market_data_scaled = pd.DataFrame(
    market_scaled_data_nparr,
    columns=["price_change_percentage_24h", "price_change_percentage_7d", "price_change_percentage_14d", 
             "price_change_percentage_30d", "price_change_percentage_60d", "price_change_percentage_200d", 
             "price_change_percentage_1y"]
)

# Copy the crypto names from the original data
df_market_data_scaled["coin_id"] = df_market_data.index

# Set the coinid column as index
df_market_data_scaled = df_market_data_scaled.set_index('coin_id')

# Display sample data
df_market_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11, where k is the name of the list
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values, where inertia is the name of the list
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters where k_model is the name of the KMeans model
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_dict = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data_dict)

df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialise the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
crypto_clusters_nparr = model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
print(crypto_clusters_nparr)
```

```python
# Create a copy of the DataFrame
df_market_data_scale_predictions = df_market_data_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_scale_predictions["CryptoCluster"] = crypto_clusters_nparr

# Display sample data
df_market_data_scale_predictions
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

df_market_data_scale_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="CryptoCluster",
    hover_cols="coin_id",
    title="Scatter Plot by Crypto Clusters"
)
```

---

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
market_data_pca_nparr = pca.fit_transform(df_market_data)

# View the first five rows of the DataFrame. 
market_data_pca_nparr[0:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.

pca.explained_variance_ratio_
```

#### Answer the following question: 
**Reference:**https://stackoverflow.com/questions/44765682/in-sklearn-decomposition-pca-why-are-components-negative

**Reference:**https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html#sklearn.decomposition.PCA

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance is the sum of the three prinical components.  
In this challenge, the first PCA (PCA1) is explaining approximately 0.97604, or almost 98% of the total variance, while the second PCA (PCA2) is explaining approximately 0.02303, or just over 2% of the total variance.  The third PCA (PCA3) is explaining approximately 0.00075, or just over 0% of the total variance. So overall, the first 2 components explain almost 100% of the variance, with the third component explaining most of the remaining variance. Therefore the total explainted variance of these 3 PCAs is approaching 1.

```python
# Create a new DataFrame with the PCA data.

# Creating a DataFrame with the PCA data
df_market_data_pca = pd.DataFrame(market_data_pca_nparr, columns=['PCA1','PCA2','PCA3'])

# Copy the crypto names from the original data
df_market_data_pca['coin_id'] = df_market_data.index

# Set the coinid column as index
df_market_data_pca = df_market_data_pca.set_index('coin_id')

# Display sample data
df_market_data_pca
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data_pca)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data_pca)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

df_elbow_pca.hvplot.line(
    x="k",
    y="inertia",
    title="PCA Elbow Curve"
).opts(yformatter="%.0f")
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialise the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state = 1)
```

```python
# Fit the K-Means model using the PCA data
model.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
crypto_clusters_4 = model.predict(df_market_data_pca)

# Print the resulting array of cluster values.
print(crypto_clusters_4)
```

```python
# Create a copy of the DataFrame with the PCA data

df_market_data_pca_predictions = df_market_data_pca.copy()

# Add a new column to the DataFrame with the predicted clusters

df_market_data_pca_predictions['CryptoCluster'] = crypto_clusters_4

# Display sample data
df_market_data_pca_predictions
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
df_market_data_pca_predictions.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="CryptoCluster",
    hover_cols="coin_id",
    title='PCA Scatter Plot by Crypto Cluster',
    width=800,
    height=400
)
```

### Visualise and Compare the Results

In this section, you will visually analyse the cluster analysis results by contrasting the outcome with and without using the optimisation techniques.

```python
# Composite plot to contrast the clusters
# Reference: https://holoviews.org/user_guide/Building_Composite_Objects.html
# Reference: https://holoviews.org/user_guide/Composing_Elements.html

import holoviews as hv
hv.extension('bokeh')

KMeans_elbow = df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)

PCA_elbow = df_elbow_pca.hvplot.line(
    x="k",
    y="inertia",
    title="PCA Elbow Curve"
)

layout = KMeans_elbow + PCA_elbow
layout
```

```python
# Composite plot to contrast the clusters

KMeans_clusters = df_market_data_scale_predictions.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="CryptoCluster",
    hover_cols="coin_id",
    title="Scatter Plot by Crypto Clusters",
    width=600,
    height=300
)

PCA_clusters = df_market_data_pca_predictions.hvplot.scatter(
    x="PCA1",
    y="PCA2",
    by="CryptoCluster",
    hover_cols="coin_id",
    title='PCA Scatter Plot by Crypto Cluster',
    width=600,
    height=300
)

layout = KMeans_clusters + PCA_clusters
layout
```

#### Answer the following question: 

  * **Question:** After visually analysing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:** KMeans shows a straight line in the Elbow Curve at approxmiately 0, and is not taking into account the effective of the many features that the PCA model analysed. The PCA model is showing the difference between the PCA1, PCA2 and PCA3 points on the elbow curve, and truly depicting how to get to k = 4 as the best number for k, having taken into account almost 100% of variances.
  * By not taking into account many features, the KMeans model has more overlap and less distinct clusters than the PCA Scatter plot. The PCA scatter plot clearly shows the 4 distinct clusters, and is clearer in demonstrating that 4 is the best number for k.
